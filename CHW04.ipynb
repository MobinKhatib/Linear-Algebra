```python
import matplotlib.pyplot as plt

# Load the BMP file
image_data = plt.imread('1.bmp')

# Display the image
plt.imshow(image_data)
plt.show()
```

```python
import numpy as np

r = image_data[:, :, 0]
g = image_data[:, :, 1]
b = image_data[:, :, 2]
```

```python
k = [10,50,150,250]  
for i in k :
    U, sig, V = np.linalg.svd(r)
    new_r = np.dot(U[:,:i], np.dot(np.diag(sig[:i]), V[:i, :]))
    U, sig, V = np.linalg.svd(g)
    new_g = np.dot(U[:, :i], np.dot(np.diag(sig[:i]), V[:i, :]))
    U, sig, V = np.linalg.svd(b)
    new_b = np.dot(U[:, :i], np.dot(np.diag(sig[:i]), V[:i, :]))
    new_image_data = np.stack((new_r, new_g, new_b), axis=2)
    plt.imshow(new_image_data)
    plt.show()
```
